```python
import numpy as np
import matplotlib.pyplot as plt

sun = np.array([0.0,0.0])
r = np.array([1.0, 0.0])
v = np.array([0.0,2*np.pi])
dt = 0.001
GM = 4*np.pi**2

def acceleration(r):
    d = np.linalg.norm(r)
    return - GM*r / d**3
trajectory = []
for _ in range(5000):
  a = acceleration(r)
  v = v + a*dt
  r = r + v*dt
  trajectory.append(r.copy())
trajectory = np.array(trajectory)
plt.plot (trajectory[:,0], trajectory[:,1])
plt.scatter (0,0, color = "orange")
plt.title ("First orbital simulation 🌍🚀")
plt.axis ("Equal")
plt.show()
```
